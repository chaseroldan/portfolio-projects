```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats as sts
from scipy.stats import linregress
```

```python
LTC_data = pd.read_csv('LTC-USD.csv')
LTC_data.head()
```

```python
BTC_data = pd.read_csv('BTC-USD.csv')
BTC_data.head()
```

```python
ETH_data = pd.read_csv('ETH-USD.csv')
ETH_data.head()
```

```python
# create data
x = LTC_data['Close']
y = BTC_data['Close']
  
# plot lines

plt.plot(x, np.sin(x), label = "curve 1", linestyle="-.")
plt.plot(x, np.cos(x), label = "curve 2", linestyle=":")
plt.legend()
plt.show()
```

```python
LTC_data.plot(kind="line", x="Date", y="Close", grid=True, figsize=(10,9),
              title="LTC Close vs. Date")
```

```python
ETH_data.plot(kind="line", x="Date", y="Close", grid=True, figsize=(10,9),
              title="ETH Close vs. Date")
```

```python

# multiple line plots

plt.plot( 'Date', 'Open', data=LTC_data, marker='', color='red', linewidth=2)
plt.plot( 'Date', 'Open', data=ETH_data, marker='', color='orange', linewidth=2)
# show legend
plt.legend()
ax1 = plt.axes()
x_axis = ax1.axes.get_xaxis()
x_axis.set_visible(False)
# show graph
plt.show()
```
