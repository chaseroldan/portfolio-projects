```python
import pandas as pd
import matplotlib as plt
```

```python
df = pd.read_csv('football.csv', header=[1]) 
df.head()
```

```python
df.info()
```

```python
# convert object to float

yds = df['YDS'].str.replace(',', '').astype(float)
```

```python
yds.head()
```

```python
# add new column to original data

df['Yards'] = yds
```

```python
df.head()
```

```python
# choose specific columns

newdf = df[['UniversityName', 'Year', 'CMP', 'ATT','TD', 'INT', 'SACK', 'TOTAL_PLAYS', 'Yards']]
newdf
```

```python
# list names of all teams in data

newdf['UniversityName'].unique()
```

```python
# scatter plot of tds vs yards

newdf.plot(kind="scatter", x="Yards", y="TD", figsize=(8,8),
              title="TD Vs. Yards")
```

```python
# specific team stats
```

```python
bama = newdf.loc[newdf['UniversityName']=='Alabama']

bama
```

```python
bama = bama.set_index('Year')
bama
```

```python
bama.describe()
```

```python
bama['Yards'].plot.bar(figsize=(8,8))
```

```python
# scatter plot of tds vs ints

bama.plot(kind="scatter", x="INT", y="TD", figsize=(8,8),
              title="TD Vs. INT")
```

```python
# scatter plot of tds vs sacks

bama.plot(kind="scatter", x="SACK", y="TD", figsize=(8,8),
              title="TD Vs. Sacks")
```

```python
html_table = bama.to_html()
html_table.replace('\n', '')
```
