```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Load in csv
df = pd.read_csv("top-songs.csv", encoding= 'unicode_escape')
df.head()
```

```python
# new data frame with split value columns 
new = df["released"].str.split(",", n = 1, expand = True) 
  
df["month"]= new[0] 
  
df["year"]= new[1] 
  
# dropping old name columns 
df.drop(columns =["released"], inplace = True) 
  
# df display 
df 
```

```python
df = df.drop_duplicates()
```

```python
df.to_csv('top500.csv')
```
