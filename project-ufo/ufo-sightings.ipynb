```python
import pandas as pd
import numpy as np
import scipy.stats as sts
from scipy.stats import linregress
```

```python
# read in data
csv_path = "ufo.csv"
ufo_df = pd.read_csv(csv_path)
ufo_df.head()
```

```python
# remove rows with missing data
clean_ufo_df = ufo_df.dropna(how="any")
clean_ufo_df.count()
```

```python
# summarize data
clean_ufo_summary = clean_ufo_df.nunique()
print(clean_ufo_summary)
```

```python
# display the df 
clean_ufo_df.head()
```

```python
# convert seconds and latitude into numeric values
converted_ufo_df = clean_ufo_df.copy()
converted_ufo_df["duration (seconds)"] = converted_ufo_df.loc[:, "duration (seconds)"].astype(float)
converted_ufo_df["latitude"] = converted_ufo_df.loc[:, "latitude"].astype(float)
converted_ufo_df.head()
```

```python
ax1 = converted_ufo_df.plot.scatter(x='latitude',
                      y='duration (seconds)',
                      c='DarkBlue')

x = converted_ufo_df.iloc[:,9]
y = converted_ufo_df.iloc[:,5]
correlation = sts.pearsonr(x, y)
print(f"The correlation between both factors is {round(correlation[0],7)}")
```

```python
# double check everything converted properly since this df will be used for the following analyses
converted_ufo_df.dtypes
```

```python
### USA Analysis ###
```

```python
# filter only USA sightings
usa_ufo_df = converted_ufo_df.loc[converted_ufo_df["country"] == "us", :]
usa_ufo_df.head()
```

```python
# group data into separate fields per state
grouped_usa_df = usa_ufo_df.groupby(['state'])

grouped_usa_df.count().head()
```

```python
# calculate number of sightings per state
state_counts = usa_ufo_df["state"].value_counts()
state_counts.head()
```

```python
# calculate sum of seconds per state
state_duration = grouped_usa_df["duration (seconds)"].sum()
state_duration.head()
```

```python
# create summary table 
state_summary_df = pd.DataFrame({"number of sightings": state_counts,
                                    "total visit time": state_duration})
state_summary_df.head()
```

```python
### International Analysis ###
```

```python
# create an international dataframe
grouped_international_data = converted_ufo_df.groupby(['country', 'state'])
grouped_international_data.count().head(10)
```

```python
# calculate sum of seconds per country and area
international_duration_df = pd.DataFrame(
    grouped_international_data["duration (seconds)"].sum())
international_duration_df.head(10)
```

```python
# list each country and area of sightings
list(international_duration_df.index.values) 
```

```python
# calculate seconds per country
country_data = converted_ufo_df.groupby(['country'])

country_duration = country_data["duration (seconds)"].sum().apply(np.ceil)
country_duration.sort_values(ascending=False)
```

```python
# calculate sightings per country

country_counts = converted_ufo_df["country"].value_counts()
country_counts.head()
```

```python
# create a summary table 
country_summary_df = pd.DataFrame({"number of sightings": country_counts,
                                    "total visit time": country_duration})
country_summary_df.sort_values(by='total visit time', ascending=False)
```
